```python
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras import layers
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from sklearn.model_selection import train_test_split
```

```python
from seaborn import load_dataset

df = load_dataset("iris")
```

```python
df["species"] = df["species"].map({"setosa": 0, "versicolor": 1, "virginica": 2})
```

```python
X, y = df.loc[:, df.columns != "species"], df["species"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, shuffle=True)
```

```python
model = keras.Sequential(
    [
        layers.Dense(units=25, activation="relu"),
        layers.Dense(units=15, activation="relu"),
        layers.Dense(units=3, activation="linear"),
    ]
)

# Adam optimizer instead of Gradient Descent

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss=SparseCategoricalCrossentropy(from_logits=True),
)
```

```python
# Round OFF Error

print(
    "Here Both Terms Should Return Same Result But First One Do Not Produce So Because Of Round off Error, Second One Is More Accurate"
)
print("1) 2 / 10000 =", 2 / 10000)

print("2) 1 + (1 / 10000) - (1 - 1 / 10000) =", 1 + (1 / 10000) - (1 - 1 / 10000))
```

```python
# Start loss will be : 0.0059
model.fit(X_test, y_test, epochs=400)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred_softmax = tf.nn.softmax(y_pred)
```

```python
tf.nn.softmax(y_pred)
```

```python
import numpy as np

array = []
for item in y_pred_softmax:
    array.append(np.argmax(item))
y_pred_array = np.array(array)
```

```python
y_pred_array
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score

print(f1_score(y_test, y_pred_array, average="macro"))
print(precision_score(y_test, y_pred_array, average="macro"))
print(recall_score(y_test, y_pred_array, average="macro"))
```
